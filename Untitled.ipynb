```python
import pandas as pd
import pymongo
import json
```

```python
def get_data():
    
    with open("local/pass.json") as file:
        file = json.loads(file.read())
        connection_string = file.get('connection_string')
        client = pymongo.MongoClient(
            connection_string)
        db = client["femicide"]
        df = pd.DataFrame(db['master'].find())
        
    return df
```

```python
df = get_data()
```
