```python
import pandas as pd
import io
import cache_magic
import numpy as np
import requests
```

```python
def filter_out_regions(df, series_code):
    df = df[df['Series Code'] == series_code]
    zimbabwe_index = np.where(df['Country Name'] == 'Zimbabwe')[0]
    zimbabwe_index = int(zimbabwe_index)
    zimbabwe_index = zimbabwe_index + 1
    df = df[:zimbabwe_index]
    
    return df
    
```

```python
df_fem_test = df_fin[df_fin['iso3'].isin(country)]
df_fem_test =df_fem_test.set_index(['iso3', 'country'])
df_fem_test = df_fem_test[df_fem_test.columns[6:]]
df_fem_test =df_fem_test.reset_index()
df_fem_test =df_fem_test.melt(id_vars = 'country')
df_fem_test =df_fem_test[1:]
df_fem_test[['year', 'type']] = df_fem_test['variable'].str.split('-', expand=True)
df_fem_test = df_fem_test.pivot(index=['country', 'year'], columns='type', values='value')
df_fem_test = df_fem_test.dropna()
```

```python
url = 'https://raw.githubusercontent.com/Tom-Whittington/Femicide_analysis/main/processed_data/original_femicide_data.csv'
download = requests.get(url).content
df_fem = pd.read_csv(io.StringIO(download.decode('utf-8')),
                dtype={'iso3_code' : 'category',
                           'Region' : 'category',
                           'Subregion' : 'category',
                           'country' : 'category',
                           'Indicator' : 'category',
                           'Disaggregation' : 'category',
                           'Gender' : 'category',
                           'Source' : 'category',
                           'Unit' : 'category',
                          'Year' : 'string'})
```

```python
df_fem_orig = df_fem[df_fem['iso3_code'].isin(country)]
df_fem_orig = df_fem_orig[df_fem_orig['Unit'] != 'Count']
df_fem_orig = df_fem_orig.loc[(df_fem_orig['Gender'] == 'Male') | (df_fem_orig['Gender'] == 'Female')]
df_fem_orig = df_fem_orig[df_fem_orig['Disaggregation'] == '-']
df_fem_orig = df_fem_orig.pivot(index=['country', 'Year'], columns='Gender', values='Value') 
```
