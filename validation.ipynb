```python
import pandas as pd
import cache_magic
import io
```

```python
df_new = pd.read_csv('test.csv')
df_new['year'] = pd.to_datetime(df_new['year'], format='%Y-%m-%d')
df_new = df_new.set_index(['iso3', 'year'])
df_new
```

```python
url = 'https://raw.githubusercontent.com/Tom-Whittington/Femicide_analysis/main/processed_data/original_femicide_data.csv'

%cache download = requests.get(url).content

df_old = pd.read_csv(io.StringIO(download.decode('utf-8')),
                     usecols =['Region',
                               'Subregion',
                               'country',
                               'iso3_code',
                               'Disaggregation',
                               'Gender',
                               'Source',
                               'Unit',
                               'Year',
                               'Value'],
                     dtype={'iso3_code' : 'category',
                           'Region' : 'category',
                           'Subregion' : 'category',
                           'country' : 'category',
                           'Gender' : 'category',
                           'Source' : 'category',
                           'Unit' : 'category',
                          'Year' : 'string'})

df_old.columns = [col.lower() for col in df_old.columns]
df_old = df_old.rename(columns={'iso3_code' : 'iso3'})
df_old['year'] = pd.to_datetime(df_old['year'], format='%Y')
df_old = df_old.set_index(['iso3', 'year'])
df_old
```

## Geographic data

```python
df_new_geo = df_new[['continent', 'region', 'country']]
df_new_geo = df_new_geo.reset_index()
df_new_geo =df_new_geo.drop(columns='year')
df_new_geo =df_new_geo.drop_duplicates()
df_new_geo
```

```python
df_old_geo = df_old[['region', 'subregion', 'country']]
df_old_geo = df_old_geo.rename(columns={'region' : 'continent', 'subregion' : 'region'})
df_old_geo = df_old_geo.reset_index()
df_old_geo =df_old_geo.drop(columns='year')
df_old_geo =df_old_geo.drop_duplicates()
df_old_geo
```

```python
df_geo = df_new_geo.merge(df_old_geo, how='outer', on='iso3', suffixes=('_new', '_old'))
df_geo = df_geo[['iso3', 'country_old', 'country_new', 'region_old', 'region_new', 'continent_old', 'continent_new']]
#df_geo.to_csv('continent_test.csv')
```

## Values

```python
#df_new_fem = df_new.copy()
df_new_fem = df_new[['femicide_count', 'femicide_rate']]
df_new_fem = df_new_fem.reset_index()
df_new_fem = df_new_fem.rename(columns={'femicide_count' : 'count', 'femicide_rate' : 'rate'})
df_new_fem['rate'] = df_new_fem['rate'].round(2)
#df_new_fem[df_new_fem['iso3'] == 'MSR']
df_new_fem
```

```python
df_old_fem = df_old[(df_old['disaggregation'] == '-') & (df_old['gender'] == 'Female')]
df_old_fem = df_old_fem.pivot(columns = 'unit', values='value')
df_old_fem = df_old_fem.rename_axis(None, axis=1)
df_old_fem = df_old_fem.reset_index()
df_old_fem = df_old_fem.rename(columns={'Count' : 'count', 'Rate per  100,000 population' : 'rate'})
df_old_fem[['count', 'rate']] = df_old_fem[['count', 'rate']].apply(pd.to_numeric, errors='coerce')
df_old_fem['rate'] = df_old_fem['rate'].round(2)
df_old_fem
```

```python
df_fem = df_new_fem.merge(df_old_fem, how='outer', on=['iso3', 'year'], suffixes=('_new', '_old'))
df_fem['count_diff'] = df_fem['count_old'] - df_fem['count_new']
df_fem['rate_diff'] = df_fem['rate_old'] - df_fem['rate_new']
df_fem = df_fem[['iso3', 'year', 'count_old', 'count_new', 'count_diff', 'rate_old', 'rate_new', 'rate_diff']]
df_fem
```
