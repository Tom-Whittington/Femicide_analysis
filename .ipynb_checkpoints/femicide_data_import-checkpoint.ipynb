```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'processed_data/original_femicide_data.csv',
                    dtype={'iso3_code' : 'category',
                           'Region' : 'category',
                           'Subregion' : 'category',
                           'country' : 'category',
                           'Indicator' : 'category',
                           'Disaggregation' : 'category',
                           'Gender' : 'category',
                           'Source' : 'category',
                           'Unit' : 'category',
                          'Year' : 'string'})
df
```

```python
df.dtypes
```

```python
df.columns = [col.lower() for col in df.columns]
df = df.rename(columns={'iso3_code' : 'iso3'})
df
```

```python
df = df[df['gender'] != 'Male']
df
```

```python
df = df[df['disaggregation'] == '-']
```

```python
df
```

```python
df = df[df['unit'] != 'Count']
df['gender'] = df['gender'].replace('Total (all ages)', 'total')
df['gender'] = df['gender'].str.lower()
df = df.sort_values(['year', 'gender', 'iso3'])
df['value'] = pd.to_numeric(df['value'], errors='coerce')
df
```

```python
df = df.pivot(index='iso3', columns=['year', 'gender'], values='value')
```

```python
df.columns = ['-'.join(col) for col in df.columns]
```

```python
df
```

```python
df_eco = pd.read_csv(r'processed_data/eco_dicot.csv', index_col='iso3')
df_eco
```

```python
df = df_eco.merge(df, how='right', left_index=True, right_index=True)
df
```

```python
df = df.dropna(subset='country')
df
```
